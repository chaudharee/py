## MINI PROJECT : Student Report Card Management System.

---

__Design and implement a Student Report Card Management System using Python that allows a teacher to:__
- Add new student records (name, roll number, subject-wise marks).
- View the report of all students.
- Display the topper(s) of the class based on average marks.
- Search for a student by roll number.
- Display all students who have failed in one or more subjects.
- Optionally update marks of any student.

```python
print("-- WELCOME TO THE STUDENT REPORT CARD MANAGER --")

student_data = {}

while True:
    print("\nChoose an operation:")
    print("[1] Add a new student")
    print("[2] View All Students")
    print("[3] Display Class Topper(s)")
    print("[4] Search Student by Roll Number")
    print("[5] Display Students Who Failed")
    print("[6] Update Student Marks")
    print("[7] Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("\n--- Add a New Student ---")
        name = input("Enter the student's name: ")
        roll_number = input("Enter the student's roll number: ")

        if roll_number in student_data:
            print("Roll number already exists. Please use a unique roll number.")
            continue
        
        print("Enter the marks out of 100 for each subject:")
        subjects = {}
        subjects['Math'] = int(input("Math marks: "))
        subjects['Science'] = int(input("Science marks: "))
        subjects['English'] = int(input("English marks: "))

        average = sum(subjects.values()) / len(subjects)

        student_data[roll_number] = {
            'name': name,
            'marks': subjects,
            'average': average
        }

        print(f"Student {name} with roll number {roll_number} added successfully.")

    elif choice == "2":
        print("\n--- View All Students ---")
        if not student_data:
            print("No student records available.")
        for roll_number in student_data:
            student = student_data[roll_number]
            print(f"Roll Number: {roll_number}, Name: {student['name']}, Marks: {student['marks']}, Average: {student['average']:.2f}")

    elif choice == "3":
        print("--- Class Topper(s) ---")
        if not student_data:
            print("No student records available.")
            continue

        highest_average = max(student['average'] for student in student_data.values())
        for roll_number in student_data:
            student = student_data[roll_number]
            if student['average'] == highest_average:
                print(f"Roll Number: {roll_number}, Name: {student['name']}, Average: {student['average']:.2f}")

    elif choice == "4":
        search_roll = input("Enter the roll number to search: ")
        if search_roll in student_data:
            student = student_data[search_roll]
            print(f"Found: Roll Number: {search_roll}, Name: {student['name']}, Marks: {student['marks']}, Average: {student['average']:.2f}")
        else:
            print("Student not found.")

    elif choice == "5":
        print("\n--- Students Who Failed ---")
        found_failed = False
        for roll_number in student_data:
            student = student_data[roll_number]
            failed_subjects = [subject for subject, mark in student['marks'].items() if mark < 40]
            if failed_subjects:
                found_failed = True
                print(f"Roll Number: {roll_number}, Name: {student['name']}, Failed Subjects: {failed_subjects}")
        if not found_failed:
            print("No students have failed in any subject.")

    elif choice == "6":
        roll_number = input("Enter the roll number of the student to update marks: ")
        if roll_number in student_data:
            print(f"Current marks for {student_data[roll_number]['name']}: {student_data[roll_number]['marks']}")
            subjects = {}
            subjects['Math'] = int(input("New Math marks: "))
            subjects['Science'] = int(input("New Science marks: "))
            subjects['English'] = int(input("New English marks: "))
            average = sum(subjects.values()) / len(subjects)
            student_data[roll_number]['marks'] = subjects
            student_data[roll_number]['average'] = average
            print(f"Marks updated for {student_data[roll_number]['name']}.")
        else:
            print("Roll number not found.")

    elif choice == "7":
        print("Exiting the program.")
        break

    else:
        print("Invalid choice. Please try again.")
```
