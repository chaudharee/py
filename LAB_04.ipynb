### __PROJECT 1__

1. __Create a CLI (Command Line Interface) contact book that allows users to:__
    - Add a new contact (append to file)
    - View all contacts (read from file)
    - Search for a contact (read and filter)
    - Handle file-related exceptions(e.g., file not found)<br><br>
File used:<br>__contacts.txt__ (stores contact info: Name, Phone)

```python
def main():
    while True:
        print("[1] To add a new contact.")
        print("[2] To view all the contact.")
        print("[3] Search for a contact.")
        print("[0] Exit.")

        choice = input("Enter your choice: ")

        if choice == "0":
            print("Exiting.....")
            exit()
        elif choice == "1":
            add_new_contact()
        elif choice == "2":
            view_all_contact()
        elif choice == "3":
            search()
        else:
            print("Invalid Input! Please enter a valid option.")

def add_new_contact():
    try:
        with open("contacts.txt", "a") as f:
            name = input("Enter the name: ")
            contact = input("Enter the phone number: ")

            f.write(f"Name : {name}, Contact: {contact}\n")
    except Exception as e:
        print("An error occured while writing in the file.")

def view_all_contact():
    try:
        with open("contacts.txt", "r") as f:
            data = f.read()
            print(data)
    except FileNotFoundError:
        print("File not found!")


def search():
    search_contact = input("Enter the contact number: ")
    try:
        with open("contacts.txt", "r") as f:
            data = f.read()
        if search_contact in data:
            print(f"{search_contact} is found in the file.")
        else:
            print("No such contact is in the file.")
    except FileNotFoundError:
        print("File not found!")

if __name__ == "__main__":
    main()
```

### __Project 2__

2. __Create a simple banking system that:__<br>
    - Strors customer info in a file
    - Allows deposits and withdrawals using functions
    - Updates customer balance
    - Logs all transactions in a separeate file
    - Handles exceptions gracefully
<br>
Files Used:<br>
**customers.txt** - stores customer records in the format:<br>
Name, AccountNumber, Balance<br>
**transactions.txt** - appends every deposit or withdrawl record with timestamp

```python
import datetime
#file names
CUSTOMERS_FILE = "customers.txt"
TRANSACTIONS_FILE = "transactions.txt"

def get_customers():
    customers = {}
    try:
        with open(CUSTOMERS_FILE, "r") as file:
            for line in file:
                name, account_number, balance = line.strip().split(',')
                customers[account_number] = {"name": name, "balance": float(balance)}
    except FileNotFoundError:
        print("Customer file not found. Starting with no customers.")
    return customers

def update_customer_file(customers):
    try:
        with open(CUSTOMERS_FILE, "w") as file:
            for acc_num, info in customers.items():
                file.write(f"{info['name']},{acc_num},{info['balance']}\n")
    except:
        IOError
        print("Erro while updating customer file!")

def log_transaction(account_number, transaction_type, amount, new_balance):
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    try:
        with open(TRANSACTIONS_FILE, "a") as file:
            file.write(f"{timestamp},{account_number},{transaction_type},{amount},{new_balance}\n")
    except IOError as e:
        print(f"Error logging transaction: {e}")

def deposit(customers):
    account_number = input("Enter account number: ")
    if account_number in customers:
        try:
            amount = float(input("Enter amount to deposit: "))
            if amount > 0:
                customers[account_number]['balance'] += amount
                print(f"Deposit successful. New balance: {customers[account_number]['balance']:.2f}")
                log_transaction(account_number, "deposit", amount, customers[account_number]['balance'])
                update_customer_file(customers)
            else:
                print("Deposit amount must be positive.")
        except ValueError:
            print("Invalid amount. Please enter a number.")
    else:
        print("Account not found.")

def withdraw(customers):
    account_number = input("Enter account number: ")
    if account_number in customers:
        try:
            amount = float(input("Enter amount to withdraw: "))
            if amount > 0 and amount <= customers[account_number]['balance']:
                customers[account_number]['balance'] -= amount
                print(f"Withdrawal successful. New balance: {customers[account_number]['balance']:.2f}")
                log_transaction(account_number, "withdrawal", amount, customers[account_number]['balance'])
                update_customer_file(customers)
            elif amount > customers[account_number]['balance']:
                print("Insufficient funds.")
            else:
                print("Withdrawal amount must be positive.")
        except ValueError:
            print("Invalid amount. Please enter a number.")
    else:
        print("Account not found.")

def banking_system():
    customers = get_customers()
    while True:
        print("\n--- Simple Banking System ---")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Exit")
        choice = input("Enter your choice: ")
        if choice == '1':
            deposit(customers)
        elif choice == '2':
            withdraw(customers)
        elif choice == '3':
            print("Exiting Banking System. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

banking_system()
```
